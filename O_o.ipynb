```python
# 1. Точечный график
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('Точечный график')
plt.show()
```

```python
# 2. Пузырьковая диаграмма с захватом группы
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
size = np.random.rand(50) * 100

plt.scatter(x, y, s=size, alpha=0.5)
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('Пузырьковая диаграмма с захватом группы')
plt.show()
```

```python
# 3. График линейной регрессии
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = 2 * x + 1 + np.random.randn(50) * 0.2

plt.scatter(x, y)
plt.plot(x, 2 * x + 1, color='red')  # Уравнение линейной регрессии
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('График линейной регрессии')
plt.show()
```

```python
# 4. Stripplot
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
plt.xlabel('День')
plt.ylabel('Сумма счета')
plt.title('Stripplot')
plt.show()
```

```python
# 5. График подсчета (Counts Plot)
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
sns.countplot(x='day', data=tips)
plt.xlabel('День')
plt.ylabel('Количество')
plt.title('График подсчета')
plt.show()
```

```python
# 6. Построчная гистограмма:
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
sns.histplot(data=tips, x='total_bill', hue='sex', multiple='stack')
plt.xlabel('Сумма счета')
plt.ylabel('Частота')
plt.title('Построчная гистограмма')
plt.show()
```

```python
# 7. Boxplot
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
sns.boxplot(x='day', y='total_bill', data=tips)
plt.xlabel('День')
plt.ylabel('Сумма счета')
plt.title('Boxplot')
plt.show()
```

```python
# 8. Диаграмма корреляции
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
corr_matrix = tips.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Диаграмма корреляции')
plt.show()
```

```python
# 9. Парный график
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
sns.pairplot(tips, hue='sex')
plt.title('Парный график')
plt.show()
```

```python
# 10. Расходящиеся столбцы
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10)
y1 = np.random.rand(10)
y2 = -np.random.rand(10)

plt.bar(x, y1, color='blue')
plt.bar(x, y2, color='red')
plt.title('Расходящиеся столбцы')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
from IPython.display import display, HTML

%matplotlib inline

x = np.arange(10)                                        # Создаем данные
y1 = np.random.rand(10)
y2 = -np.random.rand(10)

fig, ax = plt.subplots()

bar1 = ax.bar(x, y1, color='blue')                       # Создаем начальный график
bar2 = ax.bar(x, y2, color='red')

def update(frame):                                       # Функция обновления для анимации
    y1 = np.random.rand(10)
    y2 = -np.random.rand(10)
    for bar, h in zip(bar1, y1):
        bar.set_height(h)
    for bar, h in zip(bar2, y2):
        bar.set_height(h)

ani = FuncAnimation(fig, update, frames=range(10), repeat=True, interval=500)  # Создаем анимацию

display(HTML(ani.to_jshtml()))                           # Отображаем анимацию в блокноте
```
